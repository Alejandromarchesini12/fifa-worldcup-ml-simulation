# PHASE 1 - Step 3: Data Inspection and Cleaning Strategy

```python
import pandas as pd

# Loading the dataset
results = pd.read_csv('../data/raw/results.csv')
print("Results from results.csv:")
print(results.head())

rankings = pd.read_csv('../data/raw/fifa_ranking-2024-06-20.csv')
print("\n")  # Adding a newline for better readability
print("Results from fifa_ranking-2024-06-20.csv:")
print(rankings.head())
```

```python
# setting the display options to show all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 2000)
```

```python
# Displaying the info from the datasets to understand their structure
results.info()
print("\n")  # Adding a newline for better readability
print(rankings.dtypes)
```

```python
# Cleaning Column names
rankings = rankings.rename(columns={
    'country_full' : 'team',
    'rank_date' : 'date',
    'rank':'ranking',
    'total_points':'points'
})

rankings.columns
```

```python
print(results['date'].head())
print()
print(rankings['date'].head())
```

```python
# Converting Dates to datetime format
results['date'] = pd.to_datetime(results['date'])
rankings['date'] = pd.to_datetime(rankings['date'])

print(results.dtypes)
print()
print(rankings.dtypes)
```

```python
# Filter the matches that are from 1993 and onwards
results = results[results['date'] >= '1993-01-01']

results.head()
```

```python
# Inspecting the unique teams in both datasets
results_teams = set(results['home_team'].unique())
rankings_teams = set(rankings['team'].unique())

# Finding teams that are in results but not in rankings
missing_in_rankings = results_teams - rankings_teams
print("Teams in results but not in rankings:")
print(missing_in_rankings)
print(f'There are {len(missing_in_rankings)} teams in results that are not in rankings.')
```

## The Results dataset includes somes teams that are not considered FIFA teams.
This means that we need to filter. 
- We want to keep only the matches that BOTH teams exists in FIFA Rankings
- Remove all the Non-FIFA teams
- Then fix any mismatches

```python
# Identifying the Valid FIFA teams
ranking_teams = set(rankings['team'].unique())

# Filtering teams from the results
results = results[
    (results['home_team'].isin(ranking_teams)) &
    (results['away_team'].isin(ranking_teams))
]

# Checking again for the unique teams
# Inspecting the unique teams in both datasets
results_teams = set(results['home_team'].unique())
rankings_teams = set(rankings['team'].unique())

# Finding teams that are in results but not in rankings
missing_in_rankings = results_teams - rankings_teams
print("Teams in results but not in rankings:")
print(missing_in_rankings)
print(f'There are {len(missing_in_rankings)} teams in results that are not in rankings.')
```

# PHASE 1 - Step 4: Merge Rankings into Match Data

We cannot use a normal merge since the rankings change over time. 
- For each macth, we need the most recent ranking BEFORE the match date

```python
# Sorting both datasets by date
results = results.sort_values("date")
rankings = rankings.sort_values("date")

# checking the first few rows of the sorted datasets
print(results.head())
print("\n")
print(rankings.head())
```

```python
# Merging the home team rankings
home_rankings = rankings.rename(columns={
    'team': 'home_team',
    'ranking': 'home_ranking',
    'points': 'home_points'
})

results = pd.merge_asof(
    results,
    home_rankings,
    by='home_team',
    left_on='date',
    right_on='date',
    direction='backward'
)

# Merging the away team rankings
away_rankings = rankings.rename(columns={
    'team': 'away_team',
    'ranking': 'away_ranking',
    'points': 'away_points'
})

results = pd.merge_asof(
    results,
    away_rankings,
    by='away_team',
    left_on='date',
    right_on='date',
    direction='backward'
)

results.head()
```

```python
# checking for Nans
results[['home_ranking','away_ranking']].isna().sum()
```

```python
# checking percentage of Nans
print("Total matches:", len(results))
print("Home ranking NaN %:", results['home_ranking'].isna().mean() * 100)
print("Away ranking NaN %:", results['away_ranking'].isna().mean() * 100)
```

### Since the percentages of NaNs values are very low compared with the data, We are going to drop them. We would be losing less than 1% of the data.

```python
# Dropping the NaNs values
results = results.dropna(subset=['home_ranking', 'away_ranking'])

# Verifying the are dropped
print("Total matches:", len(results))
print(results[['home_ranking', 'away_ranking']].isna().sum())
```

# What we have now:
